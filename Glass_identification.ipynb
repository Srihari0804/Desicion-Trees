```python
import pandas as pd

df = pd.read_csv('/content/glass.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Type',axis=1),df['Type'], stratify=df['Type'],random_state=42,test_size=0.3)
```

```python
x_val,x_test,y_val,y_test = train_test_split(x_test,y_test,stratify=y_test,random_state=42)
```

```python
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_val = scaler.transform(x_val)
```

```python
#!pip install catboost
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(
    n_estimators=50,
    l2_leaf_reg=5,
    learning_rate= 0.1,
    loss_function='MultiClass',
    eval_metric='Accuracy',

    od_type="Iter",
    od_wait=50,
    random_state=42,

)
```

```python
model.fit(x_train,y_train,eval_set=(x_val,y_val))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,model.predict(x_test))
```

```python
r2_score(y_train,model.predict(x_train))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    max_depth=6,
    max_features=0.6,
    random_state=42,
    oob_score=True,
    verbose=1,
    max_samples=0.6
)
```

```python
model.fit(x_train,y_train)
```

```python
r2_score(y_test,model.predict(x_test))
```

```python
r2_score(y_train,model.predict(x_train))
```

```python
model = GradientBoostingClassifier(
    n_estimators= 100,
    verbose=1,
    learning_rate=0.2,
    random_state=42,
    max_depth=8
)
```
