```python
import prince
```

```python
n_components = 2
famd = prince.FAMD(
        n_components=n_components,
        n_iter=3,          # Number of iterations for the algorithm
        random_state=42    # for reproducibility
    )

# Fit FAMD to the entire DataFrame
famd = famd.fit(df)

# 3. Get the Reduced Dimensions
# We can now transform our original data into the new component space.
reduced_df = famd.transform(df)
reduced_df.columns = [f'component_{i+1}' for i in range(n_components)]

print(f"\nDataFrame with {n_components} FAMD components:")
print(reduced_df.head())

# 4. Explained Inertia
# In FAMD, "explained inertia" is the equivalent of "explained variance" in PCA.
# It tells us how much of the dataset's total variability is captured by the components.
print("\nExplained Inertia by each component:")
print(famd.explained_inertia)

total_explained_inertia = sum(famd.explained_inertia_)
print(f"\nTotal explained inertia by {n_components} components: {total_explained_inertia:.2%}")
print("\nThis means our 2 new features capture",
          f"{total_explained_inertia:.2%} of the information from all original features.")
```

###Creating Date

```python
df['Date'] = df.apply(lambda x: x['trans_date_trans_time'].split(" ")[0],axis=1)
```

```python
#Create year and month
df["Year and Month"] = df.apply(lambda x: x['Date'][0:-3],axis = 1)
```

```python
#combine first and last to get name
df['Name'] = df.apply(lambda x: x['first']+" "+x['last'],axis=1)
```

###Creating location

```python
#before creating loc lets normalize lat and long
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler2 = StandardScaler()

df.lat = scaler.fit_transform(df[['lat']])
df['long'] = scaler2.fit_transform(df[['long']])
```

```python
df['loc_of_card_holder'] = df.apply(lambda x: (x['lat'],x['long']),axis=1)
```

```python
#convert merch_loc and merch_long first
df.merch_lat = scaler.fit_transform(df[['merch_lat']])
df.merch_long = scaler.fit_transform(df[['merch_long']])
```

```python
#create merchant location
df['merch_loc'] = df.apply(lambda x: (x['merch_lat'],x['merch_long']),axis=1)
```
